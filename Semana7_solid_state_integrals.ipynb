# Cuadratura de Legendre

En física 2 siempre se asume que la capacidad calorífica de un sólido es $3 R = 3 N k_B$.

Un modelo más elaborado es el de Debye, el cual establece que cada sólido tiene una temperatura especial $T_D$, tal que su capacidad calorífica es 
$$
\tilde{C}_V(T)= \frac{C_V(T)}{N k_B} = 9 \left(\frac{T}{T_D}\right)^3 \int_0^{T_D/T} \frac{x^4 e^x }{(e^x-1)^2} \mathrm{d}x
$$

Use la cuadratura de Legendre para resolver esta integral y grafique $\tilde{C}_V$ como función de la variable $\tau = T/T_D$. Debería observar que a altas temperaturas ($\tau \approx 5$) se recupera la aproximación de física 2, mientras que a bajas temperaturas ($\tau \approx 0$) la capacidad calorífica disminuye.

Puede probar su solución con los siguientes valores:
$$
\tilde{C}_V(\tau=1/2) = 2.47622 \qquad \qquad \tilde{C}_V(\tau=1) = 2.8552
$$

# Cuadratura de Laguerre

En los conductores sólidos no sólo están los átomos del material, sino que los electrones que transportan corriente se comportan como un gas, que también tiene su contribución a la capacidad calorífica. 

La ecuación relevante en este caso es:
$$
C_e = \frac32 N k_B  \left(\frac{k_b T}{\epsilon_F}\right) \int_{-\epsilon_F / (k_B T)}^\infty \frac{x^2 e^x }{(e^x+1)^2} \mathrm{d}x
$$
Note el cambio de signo en el denominador del integrando. Defina otra variable $\tau = \frac{k_b T}{\epsilon_F}$, use la cuadratura de Laguerre para realizar esta integral, y grafique $\frac{\tilde{C}}{\tau} = \frac{2C_e}{4 N k_b \tau}$ como función de $\tau$. Debería observar que a bajas temperaturas ($\tau \approx 0$) esta capacidad calorífica es alta, y a altas temperaturas ($\tau \approx 3$) ésta se vuelve insignificante.

Puede probar con los siguientes valores:

$$
\frac{\tilde{C}}{\tau}|_{\tau=1/4} = 2.82042 \qquad \qquad \frac{\tilde{C}}{\tau}|_{\tau=1/2} =2.04332
$$
